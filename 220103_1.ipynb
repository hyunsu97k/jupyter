# Class 활용

```python
# Account Class

class Account() :
    account_cnt = 0
    
    def __init__(self, num) :
        self.__num = num
        self.__balance = 0    # __를 붙히면 외부에서 접근불가 메소드 이용해서 들어올수있음
        
        Account.account_cnt = Account.account_cnt + 1
        
    def get_num(self) :
        return self.__num
    
    def get_balance(self) :
        return self.__balance
    
    def set_num(self, num) :
        self.__num = num
        
    def set_balance(self, balance) :
        self.__balance = balance
        
    
    #입금
    def deposit(self, money) :
        self.__balance = self.__balance + money
        
    #출금    
    def withdraw(self, money) :
        
        if self.__balance >= money :
            self.__balance = self.__balance - money
        
        else :
            print('잔액이 부족합니다')
            
    #송금 !이부분 나중에 물어보기 !
    def transfer(self, money, other) :
        self.__balance = self.__balance - money
        #other.__balance = other.__balance + money 이쪽만 set이용해서 바꾸기
        
        other_balance = other.get_balance()
        other.set_balance(other_balance + money)
                
        
    #정보 출력
    def print_info(self) :
        print('계좌번호 : %s    잔액 : %d원' %(self.__num, self.__balance))
        
    #계좌 갯수 출력
    @classmethod    #데코레이션
    def print_account_cnt(cls) :
        print(cls.account_cnt)
```

## 과제 :직사각형

```python
class Rectangle :
    
    def __init__(self, garo, sero) :
        self.__garo = garo
        self.__sero = sero
    
    #def __repr__(self) : # repr --> representation
        #return 'gg'
    def __str__(self) :    # str --> string
        return '가로 : %d   세로 : %d' %(self.__garo, self.__sero)
    
    def __eq__(self, other_rect) :   # eq --> equal  == 같다는뜻
        return self.get_area() == other_rect.get_area()
    
    def __gt__(self, other_rect) : # gt --> greater than  > 크다는뜻 
        return self.get_area() > other_rect.get_area()
    
    def get_garo(self) :
        return self.__garo
    
    def set_garo(self) :
        self.__garo = garo
        
    def get_sero(self) :
        return self.__sero
    
    def set_sero(self) :
        self.__sero = sero
        
    def get_round(self) :
        #x = 2*(self.__garo + self.__sero)
        
        #return x
    
        return  2*(self.__garo + self.__sero)
    
    
    
    def get_area(self) :
        #x = self.__garo * self.__sero
        
        return self.__garo * self.__sero
    def get_diagonal(self) :       #diagonal : 대각선
        #x = (self.__garo**2 + self.__sero**2)**(1/2)
        
        return (self.__garo**2 + self.__sero**2)**(1/2)    # sqrt() --> 제곱근
```

```python
rect_list = list()

r1 = Rectangle(4, 7)
r2 = Rectangle(12, 9)
r3 = Rectangle(8, 6)
r4 = Rectangle(11, 5)
r5 = Rectangle(7, 13)

rect_list.append(r1)
rect_list.append(r2)
rect_list.append(r3)
rect_list.append(r4)
rect_list.append(r5)

print(r1)
print(r2)
print(r1 == r2)
print(r1 < r2)

#r = rect_list[0]
#r.get_area()

#rect_list[0].get_area()

for r in rect_list :
    print(r.get_area())
    
print(r2.get_diagonal())
```

```python
import json

json_string = '''[
               { "name" : "kang",
                 "age" : 26,
                 "city" : "Incheon" },
                 { "name" : "Choi",
                 "age" : 19,
                 "city" : "Seoul" }
                 ]'''                # ''' 를 쓰거나 줄바꿀때 뒤에 \ 붙히거나
        
python_list = json.loads(json_string)

print(type(python_list), python_list)
print(type(python_list[0]))

for python_dict in python_list :
    if python_dict["age"] > 20 :
        print("성인입니다.")
        
    else :
        print("미성년자 입니다.")
        
        
json_string = json.dumps(python_list)

print(type(json_string), json_string)
```
